```python
for i in range(30):
    driver.execute_script("window.scrollBy(0, 9000);")
    time.sleep(2)
```

```python
#Shirt link scrapping
shirt_link=[]
link=driver.find_elements(By.XPATH,'//a[@class="text-decoration-none text-muted text-secondary"]')
for i in link:
    shirt_link.append(i.get_attribute('href'))
    
```

```python
shirt_link
```

```python
len(shirt_link)
```

```python
ms_brand_name=[]
ms_product_name=[]
ms_price=[]
ms_product_code=[]
ms_product_type=[]
ms_category=[]
```

```python
for link in shirt_link:
    driver.get(link)
    time.sleep(2)

    
    #Finding the Product price
    try:
        price_element=driver.find_element(By.XPATH,'//div[@class="amount"]')
        if price_element:
            ms_price.append(price_element.text.replace('₹',''))
        else:
            ms_price.append('price not found')
    except Exception as e:
        ms_price.append('----')
        

        
    #Finding the Product name
    try:
        pd_element=driver.find_element(By.XPATH,'//h1[@class="prod_heading_title"]')
        if pd_element:
            ms_product_name.append(pd_element.text)
        else:
            ms_product_name.append('not found')
    except Exception as e:
        ms_product_name.append('----')
        
        
        
            #Finding the Product code
    try:
        pdc_element=driver.find_element(By.XPATH,'//p[@class="pdp-sku"]')
        if pdc_element:
            ms_product_code.append(pdc_element.text.replace('SKU : ',''))
        else:
            ms_product_code.append('Code not found')
            
    except Exception as e:
        ms_product_code.append('----')
        
        

        
     #adding brand name
    ms_brand_name.append('Fab india')
    
    #adding product type
    ms_product_type.append('Shirt')
    
                    
    #adding category
    ms_category.append('MEN')
```

```python
df1_ms=pd.DataFrame({'Brandname':ms_brand_name,'Category':ms_category,'Producttype':ms_product_type,'ProductName':ms_product_name,'Productcode':ms_product_code,'Price':ms_price})
df1_ms
```
